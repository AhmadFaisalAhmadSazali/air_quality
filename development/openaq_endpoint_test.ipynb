```python
import os

import pandas as pd
import httpx
from dotenv import load_dotenv
```

```python
load_dotenv()
key = os.getenv("OPENAQ_KEY")
headers = {"X-API-Key":key}
```

```python
base_country_url = "https://api.openaq.org/v3/countries/"
malaysia_id = "2"
malaysia_country_url = base_country_url + malaysia_id


response = httpx.get(malaysia_country_url, headers = headers)
data = response.json()
data
```

```python
# sensor data
sensor_base_url = "https://api.openaq.org/v3/sensors/{sensors_id}/days"
sensor_id = 3917
sensor_url = sensor_base_url.format(sensors_id=sensor_id)
params = {"limit":1000}

response = httpx.get(sensor_url, headers = headers, params=params)
data = response.json()
data
```

```python
results = data["results"]
```
